```python
def mapping(x):
    try:
        tmp = literal_eval(x)
    except:
        tmp=None
    return tmp
```

```python
def mapping2(x):
    try:
        tmp = literal_eval(x)
    except:
        tmp=None
    return pd.DataFrame(tmp, index=[0])
```

```python
%%time
dictionary = df['about.timezone'].apply(mapping2)
dictionary.shape
```
