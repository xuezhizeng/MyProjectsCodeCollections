```python
def one(x,y):
    result=[]
    for i in range(len(y)):
        result.append(cosine_similarity(x.values.reshape(1,-1), y.iloc[i].values.reshape(1,-1))[0][0])
        sorting = pd.Series(result).sort_values(ascending=False)
    return sorting
```
