```python
df1 = df.drop(['about.avatar', u'about.avatar_cdn', u'about.avatar_large',
       u'about.avatar_large_cdn', u'about.cover_image', u'about.id', u'userId',
                'about.corporate', 'about.preferred_freelancer', 
                ], axis=1)
```

```python
df1['Month'] = df1['about.registration_date_human'].fillna('NoMn, NoYr').apply(lambda x: x.split(',')[0] )
df1['Year'] = df1['about.registration_date_human'].fillna('NoMn, NoYr').apply(lambda x: x.split(',')[1] )

df1['about.location'] = df1['about.location'].fillna("")
```

```python
def helperD(x):
    try:
        return literal_eval(x)
    except:
        return {}
    
def select1(x):
    try: 
        return x['city']
    except:
        return ""
    
        
def select2(x):
    try: 
        return x['country']['name']
    except:
        return ""
```

```python
df1['city'] = df1['about.location'].apply(helperD).apply(select1)
df1['country'] = df1['about.location'].apply(helperD).apply(select2)
```
