```python
def helper_doctor(x):
    try:
        save = []
        lists = ast.literal_eval(x)
        for dic in lists:
            for key, value in dic.items():
                save.append((key.encode('utf-8'), value.encode('utf-8')))
        return save
    except Exception as e:
#         print e
        return x
```

```python
def helper_list(x):
    try:
        save = []
        lists = ast.literal_eval(x)
        for i in lists:
            save.append(i.encode('utf-8'))
        return save
    except Exception as e:
#         print e
        return x
```
