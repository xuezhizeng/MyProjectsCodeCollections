### an example of extracting a feature from string representation of dictionary

```python
import pandas as pd
from ast import literal_eval
```

```python
def mapping(x):
    if type(x)==str:
        
        try:
            tmp = literal_eval(x)
        except:
            tmp=None
    else:
        tmp=x
    return tmp
```

```python
%%time
df = pd.read_csv('./new_golden.csv')
print df.shape
```

```python
def retrieve(field, df=df):
    one = df[field].apply(mapping)
    two = pd.DataFrame.from_dict(one.to_dict())
    three = two.transpose()
    return three
```
