```python
def showpic(lists):
    n=len(lists)
    print 'viewing ',n, 'pics'
    
    f=0
    for i in lists:
        f+=1
        name=path+i
        if f%2==1:
            print f, '-', f+1
            plt.figure(figsize=((10,7)))
            plt.subplot(1,2,1)
            plt.imshow(plt.imread(name))
        else:

            plt.subplot(1,2,2)
            plt.imshow(plt.imread(name))
            plt.show()
```
