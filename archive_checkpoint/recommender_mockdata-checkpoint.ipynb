```python
import numpy as np
from numpy import random
random.seed(7)
```

```python
random.random()
```

```python
def genOneClaim():
    rd = [i for i in np.concatenate([random.choice(dpool, 1),random.choice(dpool, 1)*(random.random()<0.2)]) if i!='']
    # print rd

    rp = random.choice(dpool, random.randint(3,15))
    # print rp

    pool = np.concatenate([rd, rp])
    return pool
```

```python
%%time
mock=[]
for i in range(1000):
    mock.append(genOneClaim())
    
print len(mock)
```

```python
df = pd.DataFrame([{i:1 for i in j} for j in mock]).fillna(0)
df.shape
```

```python
df['index']=df.index+1
```

```python
df.head()
```

```python
dfm = df.melt(id_vars=['index'])
dfm.head()
```
