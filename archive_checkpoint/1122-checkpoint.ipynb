```python
def helper(x):
    if x<30000:
        return 'low'
    elif x<45000:
        return 'med'
    elif x<60000:
        return 'med_high'
    else:
        return 'high'
```

```python
df_cc['Income_Cut']=df_cc['Income Per Household'].apply(helper)
```
