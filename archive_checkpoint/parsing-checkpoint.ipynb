```python
def hp(x):
    try:
        return x.split(',')[-1]
    except:
        return x
```

```python
df['Mod_SubAgency']=df['SubAgency'].apply(hp)
```

```python
def split1(x):
    try:
        tmp=x.split('/')
        one=tmp[0]
        two=tmp[1]
        if len(one)>len(two):
            return one.strip()
        else:
            return two.strip()
    except:
        return x
```

```python
%%time
df['Mod_Location']=df['Location'].apply(split1)
```
