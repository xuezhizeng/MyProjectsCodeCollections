```python
def income_helper(x):
    if x<30000:
        return 'income_<30k'
    elif x<50000:
        return 'income_30k<50k'
    elif x<70000:
        return 'income_50k<70k'
    else:
        return 'income_>70k'
```

```python
def house_helper(x):
    if x<100000:
        return 'house_<100k'
    elif x<150000:
        return 'house_100k<150k'
    elif x<250000:
        return 'house_150k<250k'
    else:
        return 'house_>250k'
```

```python
df11.head()
```

```python
df11['house_val_gp']=df11['avarage_house_value'].apply(house_helper)
df11['PP_income_gp']=df11['PPP_income_househould'].apply(income_helper)
```
