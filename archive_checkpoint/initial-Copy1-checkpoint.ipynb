```python
import pandas as pd
pd.options.display.max_columns=100
pd.options.display.max_rows=1000
```

```python
df=pd.read_csv('UCONN Updated 10 Years.csv')
df.shape
```

```python
df[df['state'].isnull()].shape
```

```python
one = df[~df['state'].isnull()]
```

```python
one.shape
```

```python
one['state'] = one['state'].apply(lambda x: str(x).upper().strip().replace(',','').replace('.',''))
```

```python
one.shape
```

```python
# one['state'].value_counts()
```

```python
(one['state'].value_counts()>400).sum()
```

```python
one['state'].value_counts().count()
```

```python
save = list(one['state'].value_counts()[one['state'].value_counts()>400].index)
```

```python
two = one[one['state'].apply(lambda x: str(x) in save)]
two.shape
```

```python
two.to_csv('clean_state.csv', index=False)
```
